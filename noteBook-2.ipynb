## Removing nose
Our data cleaning process included identifying and handling outliers, which can skew statistical models. Outliers are extreme values that differ from the bulk of the data. To ensure the accuracy of our analysis, we implemented an outlier detection technique. This method effectively pinpointed potential outliers, allowing us to address them appropriately.


```python


import pandas as pd
from scipy import stats 


data = pd.read_csv('StrokeDS.csv')


column_name = 'age'


z_scores = stats.zscore(data[column_name])


threshold = 3


outliers = np.abs(z_scores) > threshold


print("Number of outliers in", column_name, ":", outliers.sum())


outlier_data = data[outliers]
print(outlier_data)
```

```python
##removing avg_glucose_level 

import numpy as np
import pandas as pd
from scipy import stats 

def detect_and_remove_outliers(data, column_name, threshold=3):

  z_scores = stats.zscore(data[column_name])

 
  outliers = np.abs(z_scores) > threshold

  return data[~outliers]


data = pd.read_csv('StrokeDS.csv')

filtered_data = detect_and_remove_outliers(data.copy(), 'avg_glucose_level')


z_scores_filtered = stats.zscore(filtered_data['avg_glucose_level'])
outliers_filtered = np.abs(z_scores_filtered) > 3 



print("\nDataFrame after outlier removal:")
print(filtered_data)
```

## Normalize Data using Min-Max Scaling:

Normalization was performed to ensure consistent data scale. The normalization technique applied is max-min normalization. This technique scales specific attribute values ​​to a specified range from 0 to 1. The following attributes were selected for normalization: age, average blood glucose level, and BMI (body mass index). can). We can use a standardized data set to provide a more consistent and comparable representation of  attributes, allowing accurate analysis and modeling to predict stroke  with specified results. determined.


```python


from sklearn.preprocessing import StandardScaler
import pandas as pd


df = pd.read_csv('StrokeDS.csv')

columns_to_normalize = ['avg_glucose_level', 'age', 'bmi']  

zscore_scaler = StandardScaler()
normalized_data_zscore = zscore_scaler.fit_transform(df[columns_to_normalize])


df[columns_to_normalize] = normalized_data_zscore

print("Z-score normalized data:")
print(df)
```

## Feature selection

we will use a feature selection tool called recursive feature elimination (RFE) to reduce the number of input variables in a predictive model. This is a widely used algorithm to select the most relevant features for prediction of the prediction target variable

```python


from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split

df = pd.read_csv('StrokeDS.csv')


X = df.drop('stroke', axis=1)  
y = df['stroke']


X = pd.get_dummies(X)


imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.3, random_state=7)


scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = LogisticRegression(max_iter=1000) 
rfe = RFE(model, n_features_to_select=2) 
X_train_selected = rfe.fit_transform(X_train_scaled, y_train)
X_test_selected = rfe.transform(X_test_scaled)

# Display the selected features
selected_features = [feature for feature, selected in zip(X.columns, rfe.support_) if selected]
print("Selected features:", selected_features)
```
